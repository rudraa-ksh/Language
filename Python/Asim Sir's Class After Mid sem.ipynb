1. Write a function to convert Binary to decimal and decimal to binary?

````python
#Recursive Function to convert Decimal to Binary

def decimalToBinary(ip_val):
    if ip_val >= 1:
    # recursive function call
        decimalToBinary(ip_val // 2)
    
    # printing remainder from each function call
    print(ip_val % 2, end = '')
 
# Driver Code
if __name__ == '__main__':
    # decimal value
    ip_val = 24
     
    # Calling special function
    decimalToBinary(ip_val)
````

Q2. Write a program to sum values of dictionary

````python
def sum_values(dict):
  sum = 0
  for value in dict.values():
    sum += value
  return sum

# Test the function
test_dict = {'a': 1, 'b': 2, 'c': 3}
print(sum_values(test_dict)) # Output: 6
````

Q3. Given a dictionary with 5 student names and their marks. Write a program that prints the
dictionary contents in the ascending order of marks.


````python
students = {
    "Alice": 75,
    "Bob": 82,
    "Charlie": 96,
    "Dave": 65,
    "Eve": 100
}

# Sort the students by their marks in ascending order
sorted_students = sorted(students.items(), key=lambda x: x[1])
# Print the sorted students and their marks
for student, mark in sorted_students:
    print(f"{student}: {mark}")
````

2.  Multilevel Inheritance

multilevel inheritance.jpg

````python
class A:
  def m1(self):
    print("parent class")

class B(A):
  def m2(self):
    print("intermediate class")

class C(B):
  def m3(self):
    print("child class")

d=C()
d.m1()
d.m2()
d.m3()
````
