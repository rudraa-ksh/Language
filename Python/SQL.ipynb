```python
def Push(Arr):
    l1=[]
    for i in Arr:
        if i%5==0:
            l1.append(i)
    if len(l1)==0:
        print("Your Stack in empty")
    else:
        print("Your Stack is: ",l1)    
Arr=eval(input("Enter a list of numbers: "))
Push(Arr)
```

```python
def Makepush(Package):
    q=int(input("Enter a package element: "))
    Package.append(q)
def Makepop(Package):
    if Package==[]:
        print("Your Stack is empty")
    else:
        print("Deleted Element: ",Package.pop())
Package=eval(input("Enter a stack: "))
Makepush(Package)
Makepop(Package)
```
